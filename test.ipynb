```python
import numpy as np
from matplotlib import pyplot as plt

width = 800
x = np.linspace(0, width, width)

signal1 = 1 + np.cos(2 * np.pi * (x / width))
signal2 = 1 + np.cos(2 * np.pi * (x / width) - 2 * np.pi / 3)
signal3 = 1 + np.cos(2 * np.pi * (x / width) - 4 * np.pi / 3)

plt.plot(signal1)
plt.plot(signal2)
plt.plot(signal3)
plt.show()
```

```python
temp = 3**0.5 * (signal2 - signal3) / (2 * signal1 - signal2 - signal3)
plt.plot(temp)
plt.show()
```

```python
fi = np.arctan(temp)
plt.plot(fi)
plt.show()
```
